# EDA GO Terms

This notebook analyzes the frequency of GO terms in the training data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Define path to data
data_path = '../../data/Train/train_terms.tsv'

# Check if file exists
if not os.path.exists(data_path):
    print(f"File not found: {data_path}")
    # Try absolute path if relative fails (fallback)
    data_path = '/home/hoangtungvum/Data/ML_UET/CAFA-6-Protein-Function-Prediction-Kaggle/data/Train/train_terms.tsv'
    if not os.path.exists(data_path):
         print(f"File still not found: {data_path}")

print(f"Loading data from: {data_path}")
```

```python
# Load data
df = pd.read_csv(data_path, sep='\t')
print(f"Loaded {len(df)} rows.")
df.head()
```

```python
# Calculate GO term frequencies
term_counts = df['term'].value_counts().sort_values(ascending=False)
print(f"Number of unique GO terms: {len(term_counts)}")
term_counts.head(20)
```

```python
term_counts[5000:5020]
```

```python
# Plot top 20 most frequent GO terms
plt.figure(figsize=(12, 8))
term_counts.head(20000).plot(kind='bar')
plt.title('Top  Most Frequent GO Terms')
plt.xlabel('GO Term')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plot top 20 most frequent GO terms
plt.figure(figsize=(12, 8))
term_counts.tail(10000).plot(kind='bar')
plt.title('Top  Most Frequent GO Terms')
plt.xlabel('GO Term')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plot top 20 most frequent GO terms
plt.figure(figsize=(12, 8))
term_counts.head(20).plot(kind='bar')
plt.title('Top 20 Most Frequent GO Terms')
plt.xlabel('GO Term')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
