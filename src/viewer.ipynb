```python
import pandas as pd
import os
```

## Load and display IA.tsv

```python
ia_filepath = os.path.join('..', 'data', 'IA.tsv')
try:
    ia_df = pd.read_csv(ia_filepath, sep='\t')
    print(f"Shape of IA.tsv: {ia_df.shape}")
    display(ia_df.head())
except FileNotFoundError:
    print(f"Error: {ia_filepath} not found. Make sure the data is in the 'data' directory.")
except Exception as e:
    print(f"An error occurred while reading IA.tsv: {e}")
```

## Load and display sample_submission.tsv

```python
sample_submission_filepath = os.path.join('..', 'data', 'sample_submission.tsv')
try:
    sample_submission_df = pd.read_csv(sample_submission_filepath, sep='\t')
    print(f"Shape of sample_submission.tsv: {sample_submission_df.shape}")
    display(sample_submission_df.head())
except FileNotFoundError:
    print(f"Error: {sample_submission_filepath} not found. Make sure the data is in the 'data' directory.")
except Exception as e:
    print(f"An error occurred while reading sample_submission.tsv: {e}")
```
